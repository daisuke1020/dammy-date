【問題2】散布図による可視化

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
arr = np.random.multivariate_normal((-3,0),((1.0,0.8),(0.8,1.0)),500)

%matplotlib inline
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("dummy")
plt.scatter(arr[:,0],arr[:,1])
plt.show()
```

【問題3】ヒストグラムによる可視化

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
arr = np.random.multivariate_normal((-3,0),((1.0,0.8),(0.8,1.0)),500)

%matplotlib inline
plt.xlabel("x1")
plt.ylabel("y")
plt.title("dummy of x1")
plt.xlim(-5.5,2.5)
plt.ylim(0,30)
plt.hist(arr[:,0],bins = 100)
plt.show()

plt.xlabel("x2")
plt.ylabel("y")
plt.title("dummy of x2")
plt.xlim(-5.5,2.5)
plt.ylim(0,30)
plt.hist(arr[:,1],bins = 100)
plt.show()
```

【問題4】データの追加

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
arr = np.random.multivariate_normal((-3,0),((1.0,0.8),(0.8,1.0)),500)
arr2 = np.random.multivariate_normal((0,-3),((1.0,0.8),(0.8,1.0)),500)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("dummy")
plt.scatter(arr[:,0],arr[:,1],label="0")
plt.scatter(arr2[:,0],arr2[:,1],label="1")
plt.legend()
plt.show()
```

【問題5】データの結合

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
arr = np.random.multivariate_normal((-3,0),((1.0,0.8),(0.8,1.0)),500)
arr2 = np.random.multivariate_normal((0,-3),((1.0,0.8),(0.8,1.0)),500)
arr0 = np.vstack([arr,arr2])
print(arr0.shape)
print(arr0)
```
